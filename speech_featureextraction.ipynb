```python
def getPitch(sound_file):
    Pitch = call(sound_file, "To Pitch", 0.0, 75, 600)
    minPitch = call(Pitch, "Get minimum", 0.0, 0.0, "Hertz", "Parabolic")
    maxPitch = call(Pitch, "Get maximum", 0.0, 0.0, "Hertz", "Parabolic")
    meanPitch = call(Pitch, "Get mean", 0.0, 0.0, "Hertz")
    sdPitch = call(Pitch, "Get standard deviation", 0.0, 0.0, "Hertz")
    Minpitch_list.append(minPitch)
    Maxpitch_list.append(maxPitch)
    Meanpitch_list.append(meanPitch)
    SdPitch_list.append(sdPitch)
```

```python
def getJitter(sound_file): 
    tpp = call(sound_file, "To PointProcess (periodic, cc)", 75.0, 600.0)
    jitter = call(tpp, "Get jitter (local)", 0.0, 0.0, 0.0001, 0.02, 1.3)
    Jitter.append(jitter)
```

```python
def getShimmer(sound_file):
    toPointProcess = call(sound_file, "To PointProcess (periodic, cc)", 75.0, 600.0)
    toShimmer = call([sound_file,toPointProcess],  "Get shimmer (local)", 0.0, 0.0, 0.0001, 0.02, 1.3, 1.6)
    Shimmer.append(toShimmer)
```

```python
def getHNR(sound_file):
    Harmonicity = call(sound_file, "To Harmonicity (cc)", 0.01, 75.0, 0.1, 1.0)
    hnr = call(Harmonicity, "Get mean", 0.0, 0.0)
    HNR.append(hnr)
```
